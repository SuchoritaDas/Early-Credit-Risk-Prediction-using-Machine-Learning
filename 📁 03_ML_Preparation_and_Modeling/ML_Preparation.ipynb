```python
df = spark.table("finance.silver_credit_risk")
display(df)
```

```python
categorical_cols = [
    "person_home_ownership",
    "loan_intent",
    "loan_grade"
]

numeric_cols = [
    "person_age",
    "person_income",
    "person_emp_length",
    "loan_amnt",
    "loan_int_rate",
    "loan_percent_income"
]
```

```python
df_clean = df.dropna(subset=categorical_cols + numeric_cols + ["loan_status"])
```

```python
from pyspark.ml.feature import StringIndexer

indexers = [
    StringIndexer(
        inputCol=c,
        outputCol=f"{c}_idx",
        handleInvalid="keep"
    )
    for c in categorical_cols
]
```

```python
from pyspark.ml.feature import VectorAssembler

assembler = VectorAssembler(
    inputCols=[f"{c}_idx" for c in categorical_cols] + numeric_cols,
    outputCol="features"
)
```

```python
from pyspark.ml import Pipeline

pipeline = Pipeline(stages=indexers + [assembler])

df_ml = pipeline.fit(df_clean).transform(df_clean)
```

```python
df_ml = df_ml.withColumnRenamed("loan_status", "label")
```

```python
%sql
DROP TABLE IF EXISTS finance.ml_ready_credit_risk;
```

```python
df_ml.printSchema()
```

```python
(
df_ml
.select("features", "label")
.write
.format("delta")
.mode("overwrite")
.saveAsTable("finance.ml_ready_credit_risk")
)
```

```python
df = spark.table("finance.ml_ready_credit_risk")
df.printSchema()
```

# Train Model

```python
train_df, test_df = df_ml.randomSplit([0.7, 0.3], seed=42)
```

```python
from pyspark.ml.classification import LogisticRegression

lr = LogisticRegression(
    featuresCol="features",
    labelCol="label",
    maxIter=20,
    regParam=0.01
)

lr_model = lr.fit(train_df)
```

```python
predictions = lr_model.transform(test_df)

predictions.select(
    "label",
    "probability",
    "prediction"
).show(5, truncate=False)
```

```python
predictions = lr_model.transform(test_df)

predictions.select(
    "label",
    "probability",
    "prediction"
).show(5, truncate=False)
```

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator

evaluator = BinaryClassificationEvaluator(
    labelCol="label",
    metricName="areaUnderROC"
)

auc = evaluator.evaluate(predictions)
auc
```
